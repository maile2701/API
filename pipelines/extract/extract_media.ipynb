```python
import os
import tempfile
from openpyxl import Workbook, load_workbook
from openpyxl.drawing.image import Image as XLImage
from PIL import Image as PILImage, ImageFile

ImageFile.LOAD_TRUNCATED_IMAGES = True

root_folder = "images"
output_path = "/Users/thanhmai/etl_pipeline test/data/Media_cleaned.xlsx"

# --- Khá»Ÿi táº¡o hoáº·c má»Ÿ file Excel ---
try:
    if os.path.exists(output_path):
        wb = load_workbook(output_path)
        ws = wb.active
        start_row = ws.max_row + 1
        print(f"ðŸ“˜ Äang cáº­p nháº­t thÃªm vÃ o file: {output_path}")
    else:
        raise FileNotFoundError
except Exception:
    print("âš ï¸ File Excel khÃ´ng há»£p lá»‡ hoáº·c chÆ°a tá»“n táº¡i â€” táº¡o má»›i.")
    wb = Workbook()
    ws = wb.active
    ws.title = "Danh sÃ¡ch áº£nh"
    ws["A1"] = "media_id"
    ws["B1"] = "event_name"
    ws["C1"] = "media"
    ws["D1"] = "media_type"
    start_row = 2

# --- Xá»­ lÃ½ áº£nh ---
row = start_row
existing_names = set()
if ws.max_row > 1:
    for r in range(2, ws.max_row + 1):
        existing_names.add(ws[f"B{r}"].value)

temp_files = []
media_counter = row - 1  # Äáº¿m media_id

for folder_name in sorted(os.listdir(root_folder)):
    folder_path = os.path.join(root_folder, folder_name)
    if not os.path.isdir(folder_path):
        continue

    print(f"ðŸ“‚ Äang xá»­ lÃ½ thÆ° má»¥c: {folder_name}")

    for file in sorted(os.listdir(folder_path)):
        if not file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.webp')):
            continue

        # Chuáº©n hÃ³a tÃªn áº£nh
        name = os.path.splitext(file)[0]
        event_name = name.replace("_", " ").strip()
        if event_name in existing_names:
            print(f"â© áº¢nh {event_name} Ä‘Ã£ cÃ³, bá» qua.")
            continue

        img_path = os.path.join(folder_path, file)

        try:
            with PILImage.open(img_path) as im:
                # Chuyá»ƒn áº£nh vá» RGB náº¿u cáº§n
                if im.mode in ("RGBA", "LA", "P"):
                    im = im.convert("RGB")
                im.thumbnail((150, 150))

                tmpf = tempfile.NamedTemporaryFile(suffix=".png", delete=False)
                tmp_path = tmpf.name
                tmpf.close()

                im.save(tmp_path)
                temp_files.append(tmp_path)

            # media_id tÄƒng dáº§n
            media_counter += 1
            media_id = f"media_{media_counter:03d}"

            # XÃ¡c Ä‘á»‹nh media_type
            event_lower = event_name.lower()
            if any(k in event_lower for k in ["cáº§u", "lá»… há»™i", "festival"]):
                media_type = "Du lá»‹ch"
            else:
                media_type = "Lá»‹ch sá»­"

            # Ghi vÃ o Excel
            ws[f"A{row}"] = media_id
            ws[f"B{row}"] = event_name
            img = XLImage(tmp_path)
            ws.add_image(img, f"C{row}")
            ws[f"D{row}"] = media_type

            existing_names.add(event_name)
            row += 1

        except Exception as e:
            print(f"âš ï¸ Lá»—i xá»­ lÃ½ áº£nh {file}: {e}")
            continue

# --- LÆ°u file Excel ---
try:
    wb.save(output_path)
    wb.close()
    print(f"âœ… ÄÃ£ lÆ°u táº¥t cáº£ áº£nh vÃ o file: {output_path}")
except PermissionError:
    print(f"âš ï¸ KhÃ´ng thá»ƒ lÆ°u vÃ¬ file '{output_path}' Ä‘ang má»Ÿ. HÃ£y Ä‘Ã³ng Excel rá»“i cháº¡y láº¡i.")

# --- XÃ³a file táº¡m ---
for tmp in temp_files:
    try:
        os.remove(tmp)
    except Exception:
        pass

print("ðŸ§¹ ÄÃ£ xÃ³a toÃ n bá»™ file táº¡m thÃ nh cÃ´ng.")
```
